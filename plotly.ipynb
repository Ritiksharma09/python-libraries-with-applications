```python
import seaborn as sns
```

```python
import plotly as p
import matplotlib.pyplot as plt
x=[1,23,45,6,7,8,9,10]
y=[2,4,6,8,36,64,32,22]
# matplotlib plot
plt.plot(x,y,marker="o")
plt.show()
```

```python
# seaborn plot
sns.lineplot(x=x,y=y,marker="o")
plt.show()
```

Plotly plot

```python
import plotly.express as px
```

```python
fig=px.line(x=x,y=y)
fig.show()
```

```python
import plotly.graph_objects as go
fig=go.Figure(go.Scatter(x=x,y=y))
fig.show()
```

titles

```python
fig=go.Figure(go.Scatter(x=x,y=y))
fig.update_layout(title="Line Graph",xaxis_title="X-axis",yaxis_title="Y-axis")
fig.show()
```

```python
x1,y1=[1,2,3,4,5],[5,4,6,7,3]
x2,y2=[1,2,3,4,5],[2,4,3,5,3]
x3,y3=[1,2,3,4,5],[6,3,12,7,11]
fig=go.Figure()
fig.add_trace(go.Scatter(x=x1,y=y1))
fig.add_trace(go.Scatter(x=x2,y=y2))
fig.add_trace(go.Scatter(x=x3,y=y3))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=x1,y=y1,name="line1",mode="lines"))
fig.add_trace(go.Scatter(x=x2,y=y2,name="line2",mode="markers"))
fig.add_trace(go.Scatter(x=x3,y=y3,name="line3",mode="lines+markers"))
fig.show()
```

BUBBLE CHARTS

```python
import plotly.graph_objects as go
```

```python
x=[1,2,3,4]
y=[4,6,8,11]
```

```python
fig=go.Figure(go.Scatter(x=x,y=y,mode="markers"))
fig.show()
```

```python
fig=go.Figure(go.Scatter(x=x,y=y,mode="markers",marker_size=[40,44,51,23]))
fig.show()
```

```python
fig=go.Figure(go.Scatter(x=x,y=y,mode="markers",marker_size=[40,6,35,60],text=['A','B','C','D']))
fig.show()
```
