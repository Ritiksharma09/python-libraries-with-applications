Matplotlib-Pie chart

to compare data we use bar graph

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=["Maths","English","Science","S.st","Hindi"]
plt.pie(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=["Maths","English","Science","S.st","Hindi"]
plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=["Maths","English","Science","S.st","Hindi"]
c=['yellow','magenta','blue','aqua','red']
plt.pie(x,labels=y,colors=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6]
y=["Maths","English","Science","S.st","Hindi"]
c=['yellow','magenta','blue','aqua','red']
plt.pie(x,labels=y,colors=c)
# to show data with color
plt.legend()
plt.show()
```
