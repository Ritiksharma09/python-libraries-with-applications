```python
df1=sns.load_dataset("penguins")
df1
```

```python

sns.barplot(x="island", y="bill_length_mm",data=df1,hue="sex")
plt.show()
```

```python
o=["Torgersen","Dream","Biscoe"]
sns.barplot(x="island", y="bill_length_mm",data=df1,hue="sex",order=o)
plt.show()
```

```python
# hue_order means changing parameters order of categorical factor i.e hue factor which is "sex" in the given data
sns.barplot(x="island", y="bill_length_mm",data=df1,hue="sex",order=o,hue_order=["Female","Male"])
plt.show()
```

```python
# coloring  without hue
sns.barplot(x="island", y="bill_length_mm",data=df1,color="red")
plt.show()
```

```python
# coloring in presence of hue or category
sns.barplot(x="island", y="bill_length_mm",data=df1,hue="sex",order=o,hue_order=["Female","Male"],palette="Accent")
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm",data=df1,hue="sex",order=o,hue_order=["Female","Male"],palette="prism")
plt.show()
```

```python
sns.barplot(x="island", y="bill_length_mm",data=df1,hue="sex",order=o,hue_order=["Female","Male"],palette="gist_heat_r")
plt.show()
```

saturation parameter

```python
sns.barplot(x="island", y="bill_length_mm",data=df1,hue="sex",order=o,hue_order=["Female","Male"],palette="Accent",saturation=1.3)
plt.show()
```

histogram in seaborn library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=sns.load_dataset("penguins")
df1
```

```python
sns.displot(df1["flipper_length_mm"]) #gives uneven distribution of data on x- axis
plt.show()
```

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240])
plt.show()
```

COLOR

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],color="red")
plt.show()
```

KDE:- KERNEL DENSITY

```python
sns.displot(df1["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,color="red")
plt.show()
```
